```python
import re
from collections import defaultdict

def get_stats(vocab):

    # Returns a dictionary of tuples representing the frequency count of pairs of characters.

    pairs = defaultdict(int)
    print(pairs)
    for word, freq in vocab.items():
        symbols = word.split()
        for i in range(len(symbols)-1):
            pairs[symbols[i],symbols[i+1]] += freq
    print()
    print(pairs)
    return pairs

def merge_vocab(pair, v_in):

    # Returns a new vocabulary with the pair of characters merged together wherever they appear.

    print("Freq Pair: ", pair)
    v_out = {}
    bigram = re.escape(' '.join(pair))
    p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
    #print(p)
    for word in v_in:
        w_out = p.sub(''.join(pair), word)
        print(w_out)
        v_out[w_out] = v_in[word]
    return v_out

def get_vocab(data):
    # Given a list of strings, returns a dictionary of words mapping to their frequency count in the data.

    vocab = defaultdict(int)

    for line in data:
        for word in line.split():
            vocab[' '.join(list(word)) + ' </w>'] += 1
    print(vocab)
    return vocab

def byte_pair_encoding(data, n):

    # Returns a list of n merged pairs of characters found in the vocabulary of the input data.

    vocab = get_vocab(data)
    for i in range(n):
        pairs = get_stats(vocab)
        best = max(pairs, key=pairs.get)
        vocab = merge_vocab(best, vocab)
    return vocab

# Example usage:
corpus = '''The biggest balloon floated across the Mississippi river.'''
data = corpus.split('.')
print(data)

n = 20
bpe_pairs = byte_pair_encoding(data, n)
bpe_pairs

```
