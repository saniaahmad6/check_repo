```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Mughal_Empire"
response = requests.get(url)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
p_tags = soup.findAll('p')
p_tags
```
